```python
import pandas as pd
from terminology_api import snomed_utils
```

```python
codelist = pd.read_csv('codelists/output/query0_opcs.csv')
```

```python
codelist
```

```python
SNOMED_SERVER = 'https://browser.ihtsdotools.org/snowstorm/snomed-ct/'
EDITION_PATH = 'MAIN'+'/'+'2023-07-31'
```

```python
len(codelist['snomed'])
len(codelist['snomed'].unique())

snomed_concepts = codelist['snomed'].unique()
```

```python
queries = ['>>' + i + '. 260686004 | Method (attribute) |' for i in snomed_concepts.astype(str)]
queries[0:10]
```

```python
query_results_methods = snomed_utils.run_ecl_queries(queries, 10000, SNOMED_SERVER+EDITION_PATH)
methods = []

for r in query_results_methods:
    try:
        df = pd.DataFrame(r['items'])
        try:
            methods.append( list(df['idAndFsnTerm']) )
        except:
            print(' error ')
            methods.append(None)
    except TypeError:
        print(' error: source concept doesn\'t exist in this SNOMED version.')
        print('        are you running this query against the same version as the original?')
```

```python
methods

methods_df = pd.DataFrame(data={
    'concept':snomed_concepts,
    'methods':methods,
    'category':None
})
display(methods_df.head())
methods_df.to_csv('codelists/output/snomed_methods.csv', index=False)


methods_dict = dict( zip( methods_df['concept'], methods_df['methods'] ))
```
